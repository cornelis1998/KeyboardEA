```python
import math
from heapq import nsmallest
import matplotlib.pyplot as plt


seed_size = 450
smallest_10_percent = []

# Divide the list into groups of size group_size
seed_groups = [pmx_uniform_01_09[x:x+seed_size] for x in range(0, len(pmx_uniform_01_09), seed_size)]

# Find lowest 10% fitness with corresponding 
for seed in seed_groups:
    num_lowest = math.ceil(len(seed) * 0.1)  
    lowest_fitness = nsmallest(num_lowest, seed, key=lambda x: x[3]) 
    lowest_fitness_COR = [(value[1], value[3]) for value in lowest_fitness]  
    smallest_10_percent.extend(lowest_fitness_COR)  

    
# Extract values
COR = [value[0] for value in smallest_10_percent]  
fitness = [value[1] for value in smallest_10_percent]  

# Create a scatter plot
plt.scatter(COR, fitness)
plt.xlabel('Index 1 Values')
plt.ylabel('Index 3 Values')
plt.title('Relationship between COR and Fitness')

plt.show()
```
